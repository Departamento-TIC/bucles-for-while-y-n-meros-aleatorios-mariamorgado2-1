````python
i = 10
while i >= 0:
    print (i)
    i -= 1
print("Despegamos!")
````

#7. Lanzador de monedas:
* Crea un programa que imprima al azar 0 ó 1.
* En lugar de 0 o 1, que imprima cara o cruz.
* Añádele un bucle para que lo haga 50 veces.
* Crea una suma acumulada para el total de veces que salió cara y las que salió cruz.


````python
import random

cara = 0
cruz = 0

print("Simulando 50 lanzamientos de moneda:")
for _ in range(50):
    resultado = random.randint(0, 1)
    if resultado == 0:
        print("Cara")
        cara += 1
    else:
        print("Cruz")
        cruz += 1

print("Resultados finales:")
print("Cara:" ,  cara)
print("Cruz:" , cruz)
````
